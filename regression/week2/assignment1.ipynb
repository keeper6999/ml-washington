```python
import pandas as pd

import numpy as np

from sklearn.linear_model import LinearRegression
```

```python
train_data = pd.read_csv('kc_house_train_data.csv')

test_data = pd.read_csv('kc_house_test_data.csv')
```

```python
train_data.describe()
```

```python
'''
bedrooms_squared = bedrooms * bedrooms
bed_bath_rooms   = bedrooms * bathrooms
log_sqft_living  = log(sqft_living)
lat_plus_long    = lat + long
'''

def transform(data):
    data['bedrooms_squared'] = data['bedrooms'] * data['bedrooms']
    data['bed_bath_rooms'] = data['bedrooms'] * data['bathrooms']
    data['log_sqft_living'] = np.log(data['sqft_living'])
    data['lat_plus_long'] = data['lat'] + data['long']
    return data
```

```python
train_data = transform(train_data)
```

```python
train_data.describe()
```

```python
def get_model(features, target):
    model = LinearRegression()
    model.fit(features, target)
    return model
```

```python
'''
Model 1: ‘sqft_living’, ‘bedrooms’, ‘bathrooms’, ‘lat’, and ‘long’
Model 2: ‘sqft_living’, ‘bedrooms’, ‘bathrooms’, ‘lat’,‘long’, and ‘bed_bath_rooms’
Model 3: ‘sqft_living’, ‘bedrooms’, ‘bathrooms’, ‘lat’,‘long’, ‘bed_bath_rooms’, ‘bedrooms_squared’, ‘log_sqft_living’, and ‘lat_plus_long’
'''
price = train_data['price']

features1 = train_data[['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long']]
model1 = get_model(features1, price)

features2 = train_data[['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long', 'bed_bath_rooms']]
model2 = get_model(features2, price)

features3 = train_data[['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long', 'bed_bath_rooms', 'bedrooms_squared', 'log_sqft_living', 'lat_plus_long']]
model3 = get_model(features3, price)
```

```python
model1.coef_
```

```python
model2.coef_
```

```python
def get_residual_sum_of_squares(model, data, real):
    predictions = model.predict(data)
    errors = real - predictions
    rss = sum(errors * errors)
    return rss
```

```python
# model1.predict(np.matrix(train_data[['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long']]))
print(get_residual_sum_of_squares(model1, features1, price))

print(get_residual_sum_of_squares(model2, features2, price))

print(get_residual_sum_of_squares(model3, features3, price))
```

```python
test_data = transform(test_data)

test_data.describe()
```

```python
test_price = test_data['price']

test_features1 = test_data[['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long']]
test_features2 = test_data[['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long', 'bed_bath_rooms']]
test_features3 = test_data[['sqft_living', 'bedrooms', 'bathrooms', 'lat', 'long', 'bed_bath_rooms', 'bedrooms_squared', 'log_sqft_living', 'lat_plus_long']]
```

```python
print(get_residual_sum_of_squares(model1, test_features1, test_price))

print(get_residual_sum_of_squares(model2, test_features2, test_price))

print(get_residual_sum_of_squares(model3, test_features3, test_price))
```
